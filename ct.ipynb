```python
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt
import os

nii_file = r"D:/ct_scan/train_fixed/train_1/train_1_a/train_1_a_1.nii.gz"

img = nib.load(nii_file)
data = img.get_fdata()

num_slices = data.shape[2]

start = int(num_slices * 0.30)
end = int(num_slices * 0.70)

output_dir = "ct_png"
os.makedirs(output_dir, exist_ok=True)

for i in range(start, end):
    slice_img = data[:, :, i]

    # Normalize to 0-255
    slice_img = (slice_img - slice_img.min()) / (
        slice_img.max() - slice_img.min() + 1e-8
    )

    plt.imsave(
        os.path.join(output_dir, f"slice_{i:03d}.png"),
        slice_img,
        cmap="gray"
    )

print(f"Saved {end-start} slices")
```
