Image extractor from folders and sub folders

```python
import os
import shutil

source_dir = r"D:\Lumora\official_data_iccv_final\files"
destination_dir = r"D:\X_ray"

os.makedirs(destination_dir, exist_ok=True)

image_extensions = ('.jpg', '.jpeg', '.png', '.bmp', '.gif', '.tiff', '.webp')

# Count images first
all_images = []

for root, dirs, files in os.walk(source_dir):
    for file in files:
        if file.lower().endswith(image_extensions):
            all_images.append(os.path.join(root, file))

print(f"Found {len(all_images)} images")

for i, source_file in enumerate(all_images, start=1):

    file = os.path.basename(source_file)
    dest_file = os.path.join(destination_dir, file)

    if os.path.exists(dest_file):
        filename, ext = os.path.splitext(file)
        counter = 1

        while os.path.exists(dest_file):
            dest_file = os.path.join(
                destination_dir,
                f"{filename}_{counter}{ext}"
            )
            counter += 1

    shutil.copy(source_file, dest_file)

    if i % 100 == 0:
        print(f"Copied {i}/{len(all_images)} images")

print("Finished!")
```

Dataset splitter between train and validation

```python
import os
import random
import shutil
from pathlib import Path

# Source dataset
SOURCE_DIR = r"D:\root_images"

# Output dataset
OUTPUT_DIR = r"D:\clf_ds"

# Split ratios
TRAIN_RATIO = 0.8
VAL_RATIO = 0.2

random.seed(42)

classes = ["xray", "others"]

for cls in classes:
    src = Path(SOURCE_DIR) / cls

    images = [
        f for f in src.iterdir()
        if f.suffix.lower() in [".jpg", ".jpeg", ".png", ".bmp", ".webp"]
    ]

    random.shuffle(images)

    train_count = int(len(images) * TRAIN_RATIO)

    train_imgs = images[:train_count]
    val_imgs = images[train_count:]

    for split, split_imgs in {
        "train": train_imgs,
        "val": val_imgs
    }.items():

        dest_dir = Path(OUTPUT_DIR) / split / cls
        dest_dir.mkdir(parents=True, exist_ok=True)

        for img in split_imgs:
            shutil.copy2(img, dest_dir / img.name)

print("Dataset split completed!")
```
